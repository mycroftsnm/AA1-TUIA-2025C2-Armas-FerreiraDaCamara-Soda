```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from sklearn.datasets import fetch_california_housing
housing = fetch_california_housing()

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler, MinMaxScaler, RobustScaler, MaxAbsScaler, PowerTransformer, QuantileTransformer
```

```python
data = pd.DataFrame(housing.data, columns=housing.feature_names)
data['PRICE'] = housing.target
```

```python
X_train, X_test, y_train, y_test = train_test_split(data.drop('PRICE', axis=1), data['PRICE'], test_size=0.2, random_state=42)
```

```python
scalers = [StandardScaler(), MinMaxScaler(), RobustScaler(), MaxAbsScaler(), PowerTransformer(), QuantileTransformer()]
```

```python
data.columns
```

```python
feature_index_1 = 1
feature_index_2 = 3
feature_name_1 = data.columns[feature_index_1]
feature_name_2 = data.columns[feature_index_2]

for scaler in scalers:
    X_train_scaled = scaler.fit_transform(X_train)
    plt.figure(figsize=(12, 6))
    plt.subplot(1, 2, 1)
    plt.title(f'Antes de escalar - {scaler.__class__.__name__}')
    plt.scatter(X_train.iloc[:, feature_index_1], X_train.iloc[:, feature_index_2])
    plt.xlabel(feature_name_1)
    plt.ylabel(feature_name_2)
    plt.subplot(1, 2, 2)
    plt.title(f'Despu√©s de escalar - {scaler.__class__.__name__}')
    plt.scatter(X_train_scaled[:, feature_index_1], X_train_scaled[:, feature_index_2])
    plt.xlabel(feature_name_1)
    plt.ylabel(feature_name_2)
    plt.tight_layout()
    plt.show()
```

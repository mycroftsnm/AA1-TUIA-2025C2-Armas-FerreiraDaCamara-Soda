```python
import numpy as np
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt
```

```python
# se crean datos sintéticos.
np.random.seed(0)
n = 100  # Número de muestras
X = np.random.rand(n, 3)  # Tres características
y = 2 * X[:, 0] + 3 * X[:, 1] + 0.5 * X[:, 2] + np.random.randn(n)  # Relación lineal con ruido
```

```python
# Dividir los datos en entrenamiento y prueba
n_train = 80
X_train, X_test = X[:n_train], X[n_train:]
y_train, y_test = y[:n_train], y[n_train:]
```

```python
### acá es donde se hace todo el análisis exploratorio de datos.
```

```python
# 1. Crear un modelo de regresión lineal de scikit-learn
lr_sklearn = LinearRegression()

# 2. Entrenar el modelo utilizando los datos de entrenamiento
lr_sklearn.fit(X_train, y_train)

# 3. Realizar predicciones en los datos de prueba
y_pred_sklearn = lr_sklearn.predict(X_test)

# 4. Calcular el error cuadrático medio (MSE) para evaluar el modelo
mse_sklearn = mean_squared_error(y_test, y_pred_sklearn)

# 5. Imprimir el MSE y visualizar las predicciones
print("MSE scikit-learn:", mse_sklearn)

# Gráfico de las predicciones
plt.scatter(y_test, y_pred_sklearn)
plt.xlabel("Valores Reales")
plt.ylabel("Predicciones")
plt.title("Regresión Lineal con scikit-learn")
plt.show()
```

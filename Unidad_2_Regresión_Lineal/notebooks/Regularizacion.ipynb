```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd


from sklearn.linear_model import LinearRegression, Ridge, Lasso, ElasticNet, RidgeCV, ElasticNetCV, LassoCV
from sklearn.model_selection import train_test_split
```

```python
np.random.seed(0)
X = np.random.rand(50, 1)
y = 2 * X + 1 + 1 * np.random.randn(50, 1)

# Crear modelos de regresi√≥n lineal, Lasso y Ridge
lr = LinearRegression()
lasso = Lasso(alpha=0.1)  # alpha controla la fuerza de la regularizaci√≥n L1 (Lasso)
ridge = Ridge(alpha=0.1)  # alpha controla la fuerza de la regularizaci√≥n L2 (Ridge)
elasticnet = ElasticNet(alpha=0.01, l1_ratio=0.5)

# Ajustar modelos a los datos
lr.fit(X, y)
lasso.fit(X, y)
ridge.fit(X, y)
elasticnet.fit(X,y)

# Mostrar coeficientes
print("Coeficientes del modelo de regresi√≥n lineal:")
print(lr.coef_)
print(lr.score(X,y))

print("\nCoeficientes del modelo Lasso:")
print(lasso.coef_)
print(lasso.score(X,y))


print("\nCoeficientes del modelo Ridge:")
print(ridge.coef_)
print(ridge.score(X,y))


print("\nCoeficientes del modelo ElasticNet:")
print(elasticnet.coef_)
print(elasticnet.score(X,y))


# Graficar resultados
plt.figure(figsize=(10, 5))
plt.scatter(X, y, label='Datos de ejemplo')
plt.plot(X, lr.predict(X), label='Regresi√≥n Lineal', color='green')
plt.plot(X, lasso.predict(X), label='Lasso', color='red')
plt.plot(X, ridge.predict(X), label='Ridge', color='blue')
plt.plot(X, elasticnet.predict(X), label='ElasticNet', color='yellow')
plt.legend()
plt.title('Comparaci√≥n de Regresi√≥n Lineal, Lasso y Ridge')
plt.show()
```

```python
# Crear modelos de regresi√≥n lineal, Lasso y Ridge
lassoCV = LassoCV()  # alpha controla la fuerza de la regularizaci√≥n L1 (Lasso)
ridgeCV = RidgeCV()  # alpha controla la fuerza de la regularizaci√≥n L2 (Ridge)
elasticnetCV = ElasticNetCV()

# Ajustar modelos a los datos
lassoCV.fit(X, y)
ridgeCV.fit(X, y)
elasticnetCV.fit(X,y)

#Mostrar alphas
print(lassoCV.alpha_)
print(ridgeCV.alpha_)
print(elasticnetCV.alpha_ )

# Mostrar coeficientes
print("\nCoeficientes del modelo Lasso:")
print(lassoCV.coef_)
print(lassoCV.score(X,y))


print("\nCoeficientes del modelo Ridge:")
print(ridgeCV.coef_)
print(ridgeCV.score(X,y))


print("\nCoeficientes del modelo ElasticNet:")
print(elasticnetCV.coef_)
print(elasticnetCV.score(X,y))


# Graficar resultados
plt.figure(figsize=(10, 5))
plt.scatter(X, y, label='Datos de ejemplo')
plt.plot(X, lassoCV.predict(X), label='Lasso', color='red')
plt.plot(X, ridgeCV.predict(X), label='Ridge', color='blue')
plt.plot(X, elasticnetCV.predict(X), label='ElasticNet', color='yellow')
plt.legend()
plt.title('Comparaci√≥n de Regresi√≥n Lineal, Lasso y Ridge')
plt.show()
```
